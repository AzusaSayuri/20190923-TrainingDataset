```python
import pandas as pd
import numpy as np
```

```python
df_master=pd.read_excel('20190920_MO_comparison_master.xlsx')
df_sub=pd.read_excel('20190920_MO_comparison_subset.xlsx')
```

```python
df_master.head()
```

```python
df_sub.head()
```

```python
col=['master mass','sub mass','master rt','sub rt','master intensity','sub intensity']
df_comp=pd.DataFrame(columns=col)
df_comp
```

```python
df_sub.index
```

```python
if df_sub.loc[1,'mass']-df_master.loc[2,'mass']<(df_sub.loc[1,'mass']* 0.00001):
    print ('1')
else :
    print('0')
```

```python
a=0
for i in df_sub.index:
    for j in df_master.index:
        if abs(df_sub.loc[i,'rt']-df_master.loc[j,'rt'])<0.1 and abs(df_sub.loc[i,'mass']-df_master.loc[j,'mass'])<(df_sub.loc[i,'mass']* 0.00001):
            df_comp.loc[a,'master mass']=df_master.loc[j,'mass']
            df_comp.loc[a,'sub mass']=df_sub.loc[i,'mass']
            df_comp.loc[a,'master rt']=df_master.loc[j,'rt']
            df_comp.loc[a,'sub rt']=df_sub.loc[i,'rt']
            df_comp.loc[a,'master intensity']=df_master.loc[j,'intensity']
            df_comp.loc[a,'sub intensity']=df_sub.loc[i,'intensity']
            a=a+1
            break
    
```

```python
df_comp['delta mass']=df_comp['master mass']-df_comp['sub mass']
df_comp['delta rt']=df_comp['master rt']-df_comp['sub rt']
df_comp['delta intensity']=df_comp['master intensity']-df_comp['sub intensity']
df_comp.head()
```

```python
df_comp[(abs(df_comp['delta mass'])>0.05) | (abs(df_comp['delta rt'])>0.1)]
```

```python
df_merge=df_comp[(abs(df_comp['delta mass'])<0.05) | (abs(df_comp['delta rt'])<0.1)]
df_merge.head()
```

```python
df_conc=df_merge[abs(df_merge['delta intensity'])>1e5]
```

```python
print(df_master.shape[0],'in master file,',df_sub.shape[0],'in sub file',df_merge.shape[0],'features were shared through out two files,',df_conc.shape[0],'features could be a different peak')
```
